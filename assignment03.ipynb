## Task 1

Develop an algorithm that would sort numbers from least to greatest. Specifically, given an unsorted set of N decimal values, your function should sort them to give an answer of the sorted data. For example, if you get the list of numbers [2.1, 0, 5.7, -1.2, 3, 1.5] as input, your function should return [-1.2, 0, 1.5, 2.1, 3, 5.7].

You are **NOT** allowed to use sort function available in any package!

```python
def sort(input_list):
    ss = input_list[0]
    bs = 0
    l1 = []
    l2 = []
    ilc = input_list
    for i in range(len(input_list)):
        for item in ilc:
            if(ss>item):
                ss = item
            if(item>bs):
                bs = item
        l1.append(ss)
        l2.append(bs)
        if(ss in ilc):
            ilc.remove(ss)
        if(bs in ilc):
            ilc.remove(bs)
        if(ilc==[]):
            break
        ss = ilc[0]
        bs = 0
    output_list = l1 + l2[::-1]
    return output_list
```

```python
print(sort([2.1, 0, 5.7, -1.2, 3, 1.5] ))
```

## Task 2

Write a function which gets a non-negative integer as input and prints a triangle with the height of given number. For example, given an input 3, the function should print:

    *
    **
    ***
    
**Hint**: read the documentation for [print](https://docs.python.org/3/library/functions.html#print) function. You can specify the *end* argument.

```python
def star(n):
    star = '*'
    for i in range(0,n):
        yield star
        star+='*'
```
